### Dataset Structure Overview Columns:

- order_id: Unique identifier for each order.
- customer_id: Unique identifier for each customer.
- order_date: Date when the order was placed (past 2 years).
- delivery_date: Date when the order was delivered (past 2 years).
- product_id: Unique identifier for each product.
- product_name: Name of the product (e.g., Cargo Shipping, Quick).
- price: Product price (₹50 to ₹2000).
- units: Quantity purchased (1 to 25).
- discount: Discount applied (₹0 to ₹50).
- payment_method: Payment type (e.g., Credit Card, UPI).
- order_status: Status of the order (e.g., Delivered, Cancelled).
- shipping_cost: Shipping cost (units × ₹26).
- delivery_time: Delivery time in days (1 to 14).
- customer_rating: Customer rating (1 to 5).
- region: Geographic region (e.g., North, South).
- is_returned: Return flag (0 or 1).
- promotion_code: Applied promo code (e.g., PROMO10).
- device_type: Device used for order (e.g., Mobile).
- customer_tier: Membership tier (e.g., Regular, Gold).
- vendor_id: Unique identifier for each vendor.
- vendor_rating: Vendor rating (2.5 to 5.0).
- stock_available: Stock availability flag (0 or 1).
- total_sales_value: Total sale value (₹50 to ₹2000).
- traffic_source: Source of traffic (e.g., Organic, Paid).
- product_category: Product category (e.g., Shipping, CROSS-BORDER).
- description: Product description based on category.

`Rs. 26/500gms`

- Example:

- If your package weighs 1200 grams, it will be rounded up to 1500 grams (3 units of 500 grams). The shipping cost would be:

`3 units * Rs. 26/unit = Rs. 78`

### Feature Engineering & Preprocessing

```python
# Find the mode of the 'promotion_code' column
mode_value = df['promotion_code'].mode()[0]

# Replace null values with the mode
df['promotion_code'].fillna(mode_value, inplace=True)
```

```python
# Mapping for product_name to multiple product_categories and descriptions

# Product Name mapping
product_name_mapping = {
    'CargoX': '1',
    'Quick': '2',
    'ShipX': '3',
    'LaunchX': '4',
    'Cargo Shipping': '5',
    'Fulfillment': '5'
}
df["product_name"] = df["product_name"].map(product_name_mapping)

# Product Category mapping
product_category_mapping = {
    'CROSS-BORDER': '1',
    'Shipping': '2'
}
df["product_category"] = df["product_category"].map(product_category_mapping)

# Product description mapping
description_mapping = {
    'Cross-border B2B & bulk shipping': '1',
    'Speedy local deliveries at low costs': '2',
    'Shipping to 220+ countries & territories': '3',
    'End-to-end unified retail enablement solution': '4',
    'Nationwide B2B & bulk shipping': '5',
    'Warehousing, packaging & shipping services': '6'
}
df["description"] = df["description"].map(description_mapping)

# Payment method mapping
payment_method_mapping = {
    'Credit Card': '1',
    'Debit Card': '2',
    'Net Banking': '3',
    'UPI': '4',
    'Cash on Delivery': '5'
}
df["payment_method"] = df["payment_method"].map(payment_method_mapping)

# Order status mapping
order_status_mapping = {
    'Delivered': '1',
    'Cancelled': '0',
    'Returned': '2',
    'Pending': '3'
}
df["order_status"] = df["order_status"].map(order_status_mapping)

# Region mapping
region_mapping = {
    'North': '1',
    'South': '2',
    'East': '3',
    'West': '4',
    'Central': '5'
}
df["region"] = df["region"].map(region_mapping)

# Device type mapping
device_type_mapping = {
    'Mobile': '1',
    'Desktop': '3',
    'Tablet': '2'
}
df["device_type"] = df["device_type"].map(device_type_mapping)

# Customer tier mapping
customer_tier_mapping = {
    'Regular': '1',
    'Silver': '2',
    'Gold': '3',
    'Platinum': '4'
}
df["customer_tier"] = df["customer_tier"].map(customer_tier_mapping)

# Traffic source mapping
traffic_source_mapping = {
    'Organic': '1',
    'Paid': '2',
    'Referral': '3',
    'Social Media': '4'
}
df["traffic_source"] = df["traffic_source"].map(traffic_source_mapping)

# Promotion code mapping
promotion_code_mapping = {
    'PROMO10': '10',
    'DISCOUNT20': '20',
    'SAVE15': '15'
}
df["promotion_code"] = df["promotion_code"].map(promotion_code_mapping)
```
