```python
pd.read_sql_query('SELECT * FROM shiprocketdb.shiprocket_dataset LIMIT 5',connection)
```

```python
# Total Orders and Customers

pd.read_sql('''
SELECT COUNT(order_id) AS total_orders, COUNT(DISTINCT customer_id) AS unique_customers
FROM shiprocket_dataset;

''', connection) 
```

```python
# Top 5 Region with the Most Orders

pd.read_sql_query('''
SELECT region, COUNT(order_id) AS total_orders
FROM shiprocket_dataset
GROUP BY region
ORDER BY total_orders DESC
LIMIT 5;

''', connection)
```

```python
# Average Delivery Time

pd.read_sql_query('''
SELECT AVG(TIMESTAMPDIFF(SECOND, STR_TO_DATE(order_date, '%Y-%m-%d'), delivery_time) / (60 * 60 * 24)) AS avg_delivery_time_days
FROM shiprocket_dataset;
''', connection)
```

```python
# Orders Delivered Late

pd.read_sql_query('''
SELECT COUNT(order_id) AS late_orders
FROM shiprocket_dataset
WHERE delivery_time > 12;

''', connection)
```

```python
# Sales Over Time

pd.read_sql_query('''
SELECT DATE(order_date) AS order_date, SUM(total_sales_value) AS total_sales
FROM shiprocket_dataset
GROUP BY order_date
ORDER BY order_date LIMIT 30;

''', connection)
```

```python
# Top Delivery Personnel by Efficiency

pd.read_sql_query('''
SELECT order_id, AVG(TIMESTAMPDIFF(MINUTE, order_date, delivery_time)) AS avg_delivery_time_minutes
FROM shiprocket_dataset
GROUP BY order_id
ORDER BY avg_delivery_time_minutes ASC
LIMIT 5;

''', connection)
```

```python
# Sales Growth Over Time

pd.read_sql_query('''
SELECT 
    DATE_FORMAT(order_date, '%Y-%m') AS order_month,
    SUM(total_sales_value) AS total_sales,
    LAG(SUM(total_sales_value)) OVER (ORDER BY DATE_FORMAT(order_date, '%Y-%m')) AS previous_sales,
    (SUM(total_sales_value) - LAG(SUM(total_sales_value)) OVER (ORDER BY DATE_FORMAT(order_date, '%Y-%m'))) / 
    LAG(SUM(total_sales_value)) OVER (ORDER BY DATE_FORMAT(order_date, '%Y-%m')) * 100 AS growth_percentage
FROM shiprocket_dataset
GROUP BY order_month
ORDER BY order_month;

''', connection)
```

```python
# Customer Segmentation

pd.read_sql_query('''
SELECT 
    customer_id,
    SUM(total_sales_value) AS total_spent,
    CASE
        WHEN SUM(total_sales_value) < 200 THEN 'Low Spender'
        WHEN SUM(total_sales_value) BETWEEN 500 AND 1000 THEN 'Medium Spender'
        ELSE 'High Spender'
    END AS spending_category
FROM shiprocket_dataset
GROUP BY customer_id
ORDER BY total_spent DESC LIMIT 30;

''', connection)
```

```python
# Delivery Efficiency by Distance

pd.read_sql_query('''
SELECT 
    delivery_time,
    AVG(TIMESTAMPDIFF(MINUTE, order_date, delivery_time)) AS avg_delivery_time,
    COUNT(order_id) AS total_orders
FROM shiprocket_dataset
GROUP BY delivery_time
ORDER BY delivery_time;

''', connection)
```

```python
# Revenue Contribution by Product Categories

pd.read_sql_query('''
SELECT 
    product_category,
    SUM(total_sales_value) AS total_revenue,
    (SUM(total_sales_value) / (SELECT SUM(total_sales_value) FROM shiprocket_dataset) * 100) AS revenue_percentage
FROM shiprocket_dataset
GROUP BY product_category
ORDER BY total_revenue DESC;

''', connection)
```

```python
# Delivery Personnel Performance

pd.read_sql_query('''
SELECT
    order_id,  -- Or the actual column name for delivery person ID
    COUNT(order_id) AS total_deliveries,
    AVG(TIMESTAMPDIFF(MINUTE, order_date, delivery_time)) AS avg_delivery_time_minutes,
    SUM(CASE WHEN TIMESTAMPDIFF(MINUTE, order_date, delivery_time) <= 0 THEN 1 ELSE 0 END) AS on_time_deliveries,  -- Corrected on-time calculation
    (SUM(CASE WHEN TIMESTAMPDIFF(MINUTE, order_date, delivery_time) <= 0 THEN 1 ELSE 0 END) / COUNT(order_id)) * 100 AS on_time_percentage
FROM shiprocket_dataset
GROUP BY order_id  -- Group by delivery person
ORDER BY on_time_percentage DESC, avg_delivery_time_minutes ASC
LIMIT 10;

''', connection)
```

```python
# Revenue by Day of the Week

pd.read_sql_query('''
SELECT 
    DAYNAME(order_date) AS day_of_week,
    SUM(total_sales_value) AS total_sales,
    COUNT(order_id) AS total_orders
FROM shiprocket_dataset
GROUP BY day_of_week
ORDER BY FIELD(day_of_week, 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday');

''', connection)
```

```python
# Monthly Sales Trends

pd.read_sql_query('''
SELECT 
    DATE_FORMAT(order_date, '%Y-%m') AS month,
    COUNT(order_id) AS total_orders,
    SUM(total_sales_value) AS total_sales,
    AVG(total_sales_value) AS avg_order_value
FROM shiprocket_dataset
GROUP BY month
ORDER BY month;

''', connection)
```

```python
# Geographical Insights

pd.read_sql_query('''
SELECT 
    vendor_id,
    COUNT(order_id) AS total_orders,
    SUM(total_sales_value) AS total_revenue
FROM shiprocket_dataset
GROUP BY vendor_id
ORDER BY total_revenue DESC LIMIT 5;

''', connection)
```

```python
pd.read_sql_query('''
SELECT
    vendor_id,
    COUNT(order_id) AS total_deliveries,
    SUM(CASE WHEN DATE(order_date) = DATE(delivery_time) THEN 24 ELSE 12 END) AS on_time_deliveries,
    (SUM(CASE WHEN DATE(order_date) = DATE(delivery_time) THEN 24 ELSE 12 END) / COUNT(order_id)) * 100 AS on_time_percentage
FROM shiprocket_dataset
GROUP BY vendor_id
ORDER BY on_time_percentage DESC
LIMIT 10;

''', connection)
```

```python
# Revenue Contribution by Product Category

pd.read_sql_query('''
SELECT 
    product_category,
    SUM(total_sales_value) AS total_revenue,
    (SUM(total_sales_value) / (SELECT SUM(total_sales_value) FROM shiprocket_dataset)) * 100 AS revenue_percentage
FROM shiprocket_dataset
GROUP BY product_category
ORDER BY total_revenue DESC;

''', connection)
```

```python
# Revenue Contribution by Product Name

pd.read_sql_query('''
SELECT 
    product_name,
    SUM(total_sales_value) AS total_revenue,
    (SUM(total_sales_value) / (SELECT SUM(total_sales_value) FROM shiprocket_dataset)) * 100 AS revenue_percentage
FROM shiprocket_dataset
GROUP BY product_name
ORDER BY total_revenue DESC;

''', connection)
```

```python
# Weekly Orders Performance

pd.read_sql_query('''
SELECT 
    DAYNAME(order_date) AS day_of_week,
    SUM(total_sales_value) AS total_revenue,
    COUNT(order_id) AS total_orders
FROM shiprocket_dataset
GROUP BY day_of_week
ORDER BY FIELD(day_of_week, 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday');

''', connection)
```

```python
# Monthly Order Status Analysis

pd.read_sql_query('''
SELECT 
    DATE_FORMAT(order_date, '%Y-%m') AS month,
    order_status,
    COUNT(order_id) AS total_orders
FROM shiprocket_dataset
GROUP BY month, order_status
ORDER BY month DESC, total_orders DESC LIMIT 30;

''', connection)
```

```python
# Monthly Order Status with Product Name & Category

pd.read_sql_query('''
SELECT 
    DATE_FORMAT(order_date, '%Y-%m') AS month,
    product_name,
    product_category,
    order_status,
    COUNT(order_id) AS total_orders
FROM shiprocket_dataset
GROUP BY month, product_name, product_category, order_status
ORDER BY month DESC, total_orders DESC LIMIT 30;

''', connection)
```

```python
# Total Orders and Revenue

pd.read_sql_query('''
SELECT COUNT(order_id) AS total_orders, 
       SUM(total_sales_value) AS total_revenue
FROM shiprocket_dataset;

''', connection)
```

```python
# Monthly Order Trends

pd.read_sql_query('''
SELECT DATE_FORMAT(order_date, '%Y-%m') AS order_month,
       COUNT(order_id) AS total_orders,
       SUM(total_sales_value) AS revenue
FROM shiprocket_dataset
GROUP BY order_month
ORDER BY order_month;

''', connection)
```

```python
# Average Delivery Time by Region

pd.read_sql_query('''
SELECT region, 
       AVG(delivery_time) AS avg_delivery_days
FROM shiprocket_dataset
GROUP BY region
ORDER BY avg_delivery_days ASC;

''', connection)
```

```python
# Top 5 Customers by Purchase Amount

pd.read_sql_query('''
SELECT customer_id, 
       SUM(total_sales_value) AS total_spent
FROM shiprocket_dataset
GROUP BY customer_id
ORDER BY total_spent DESC
LIMIT 5;

''', connection)
```

```python
# Vendor Performance Analysis

pd.read_sql_query('''
SELECT vendor_id, 
       COUNT(order_id) AS orders_handled, 
       AVG(vendor_rating) AS avg_rating
FROM shiprocket_dataset
GROUP BY vendor_id
ORDER BY avg_rating DESC
LIMIT 10;

''', connection)
```

```python
# Orders with Delayed Deliveries

pd.read_sql_query('''
SELECT order_id, 
       order_date, 
       delivery_date, 
       (JULIANDAY(delivery_date) - JULIANDAY(order_date)) AS actual_delivery_days,
       delivery_time AS expected_delivery_days
FROM shiprocket_dataset
WHERE (JULIANDAY(delivery_date) - JULIANDAY(order_date)) > delivery_time;

''', connection)
```

```python
# Payment Method Distribution

pd.read_sql_query('''
SELECT payment_method, 
       COUNT(order_id) AS total_orders
FROM shiprocket_dataset
GROUP BY payment_method
ORDER BY total_orders DESC;

''', connection)
```

```python
# Customer Ratings by Product Category

pd.read_sql_query('''
SELECT product_category, 
       AVG(customer_rating) AS avg_rating
FROM shiprocket_dataset
GROUP BY product_category
ORDER BY avg_rating DESC;

''', connection)
```

```python
# Monthly Sales Performance

pd.read_sql_query('''
SELECT DATE_FORMAT(order_date, '%Y-%m') AS order_month,
       COUNT(order_id) AS total_orders,
       SUM(total_sales_value) AS total_revenue
FROM shiprocket_dataset
GROUP BY order_month
ORDER BY order_month DESC;

''', connection)
```

```python
# Top 5 Best-Selling Products

pd.read_sql_query('''
SELECT product_name, 
       COUNT(order_id) AS total_orders, 
       SUM(units) AS total_units_sold, 
       SUM(total_sales_value) AS revenue_generated
FROM shiprocket_dataset
GROUP BY product_name
ORDER BY revenue_generated DESC
LIMIT 5;

''', connection)
```

```python
# Region-wise Order Distribution

pd.read_sql_query('''
SELECT C.region, 
       COUNT(S.order_id) AS total_orders, 
       SUM(S.total_sales_value) AS revenue
FROM shiprocket_dataset S
JOIN shiprocket_dataset C ON S.customer_id = C.customer_id
GROUP BY C.region
ORDER BY revenue DESC;

''', connection)
```

```python
# Fastest vs. Slowest Vendors (Delivery Time)

pd.read_sql_query('''
SELECT V.vendor_id, 
       COUNT(O.order_id) AS orders_handled, 
       AVG(O.delivery_time) AS avg_delivery_days
FROM shiprocket_dataset O
JOIN shiprocket_dataset V ON O.vendor_id = V.vendor_id
GROUP BY V.vendor_id
ORDER BY avg_delivery_days ASC
LIMIT 5;

''', connection)
```

```python
# Orders Delivered Late vs. On Time

pd.read_sql_query('''
SELECT
    CASE
        WHEN TIMESTAMPDIFF(DAY, order_date, delivery_date) > delivery_time
        THEN 'Late Delivery'
        ELSE 'On Time'
    END AS delivery_status,
    COUNT(order_status) AS order_count
FROM shiprocket_dataset
GROUP BY delivery_status;

''', connection)
```

```python
pd.read_sql_query('''
SELECT
    CASE
        WHEN TIMESTAMPDIFF(DAY, order_date, delivery_date) > delivery_time
        THEN 'Late Delivery'
        ELSE 'On Time'
    END AS delivery_status,
    COUNT(order_id) AS order_count,
    COUNT(product_category) AS count_product_categories, 
    COUNT(product_name) AS count_product_names           
FROM shiprocket_dataset
GROUP BY delivery_status;

''', connection)
```

```python
# Customer Lifetime Value (CLV)

pd.read_sql_query('''
SELECT C.customer_id, 
       COUNT(O.order_id) AS total_orders, 
       SUM(O.total_sales_value) AS total_spent, 
       AVG(O.total_sales_value) AS avg_order_value
FROM shiprocket_dataset O
JOIN shiprocket_dataset C ON O.customer_id = C.customer_id
GROUP BY C.customer_id
ORDER BY total_spent DESC
LIMIT 10;

''', connection)
```

```python
# Return Rate by Product Category

pd.read_sql_query('''
SELECT product_category, 
       COUNT(CASE WHEN is_returned = 1 THEN order_id END) AS returned_orders, 
       COUNT(order_id) AS total_orders, 
       ROUND( (COUNT(CASE WHEN is_returned = 1 THEN order_id END) * 100.0) / COUNT(order_id), 2) AS return_rate
FROM shiprocket_dataset
GROUP BY product_category
ORDER BY return_rate DESC;

''', connection)
```

```python
# Return Rate by Product Name

pd.read_sql_query('''
SELECT product_name, 
       COUNT(CASE WHEN is_returned = 1 THEN order_id END) AS returned_orders, 
       COUNT(order_id) AS total_orders, 
       ROUND( (COUNT(CASE WHEN is_returned = 1 THEN order_id END) * 100.0) / COUNT(order_id), 2) AS return_rate
FROM shiprocket_dataset
GROUP BY product_name
ORDER BY return_rate DESC;

''', connection)
```

```python
pd.read_sql_query('''
SELECT
    product_category,
    COUNT(CASE WHEN order_status = 'Returned' THEN order_id END) AS returned_orders,
    COUNT(CASE WHEN order_status = 'Pending' THEN order_id END) AS pending_orders,
    COUNT(CASE WHEN order_status = 'Cancelled' THEN order_id END) AS cancelled_orders,
    COUNT(CASE WHEN order_status = 'Delivered' THEN order_id END) AS delivered_orders,
    COUNT(order_id) AS total_orders,
    ROUND((COUNT(CASE WHEN order_status = 'Returned' THEN order_id END) * 100.0) / COUNT(order_id), 2) AS return_rate,
    COUNT(DISTINCT product_name) AS distinct_product_names
FROM shiprocket_dataset
GROUP BY product_category
ORDER BY return_rate DESC;

''', connection)
```

```python
# Most Effective Marketing Channels

pd.read_sql_query('''
SELECT traffic_source, 
       COUNT(order_id) AS total_orders, 
       SUM(total_sales_value) AS revenue
FROM shiprocket_dataset
GROUP BY traffic_source
ORDER BY revenue DESC;

''', connection)
```

```python
# Predicting Potential Late Deliveries

pd.read_sql_query('''
SELECT order_id, 
       order_date, 
       delivery_date, 
       delivery_time,
       TIMESTAMPDIFF(DAY, order_date, delivery_date) AS delivery_time,
       CASE
           WHEN TIMESTAMPDIFF(DAY, order_date, delivery_date) > delivery_time
           THEN 'Potential Delay'
           ELSE 'On Time'
       END AS delivery_status
FROM shiprocket_dataset LIMIT 30;

''', connection)
```

```python
# Best & Worst Performing Vendors (Ratings)

pd.read_sql_query('''
SELECT V.vendor_id, 
       COUNT(O.order_id) AS total_orders, 
       AVG(V.vendor_rating) AS avg_rating
FROM shiprocket_dataset O
JOIN shiprocket_dataset V ON O.vendor_id = V.vendor_id
GROUP BY V.vendor_id
ORDER BY avg_rating DESC
LIMIT 10;

''', connection)
```

```python
# Monthly Order Status Breakdown

pd.read_sql_query('''
SELECT
    DATE_FORMAT(order_date, '%Y-%m') AS order_month,
    order_status,
    COUNT(order_id) AS total_orders
FROM shiprocket_dataset
GROUP BY order_month, order_status
ORDER BY order_month DESC, total_orders DESC LIMIT 30;

''', connection)
```

```python
# Impact of Discounts on Sales

pd.read_sql_query('''
SELECT 
    discount, 
    COUNT(order_id) AS order_count, 
    AVG(total_sales_value) AS avg_order_value, 
    SUM(total_sales_value) AS total_revenue
FROM shiprocket_dataset
GROUP BY discount
ORDER BY total_revenue DESC LIMIT 10;

''', connection)
```

```python
# Shipping Cost and Order Status as a Percentage of Revenue

pd.read_sql_query('''
SELECT 
    order_status,
    region,
    shipping_cost,
    total_sales_value,
    (shipping_cost / total_sales_value) * 100 AS shipping_cost_percentage
FROM shiprocket_dataset
WHERE total_sales_value > 0
ORDER BY shipping_cost_percentage DESC LIMIT 10;

''', connection)
```

```python
# Incomplete Delivery Loss

pd.read_sql_query('''
SELECT 
    order_status,
    COUNT(order_id) AS total_orders,
    SUM(total_sales_value) AS revenue_loss
FROM shiprocket_dataset
WHERE order_status IN ('Returned', 'Pending', 'Cancelled')
GROUP BY order_status
ORDER BY revenue_loss DESC;

''', connection)
```

```python
# Monthly Incomplete Orders & Revenue Loss

pd.read_sql_query('''
SELECT 
    DATE_FORMAT(order_date, '%Y-%m') AS month,
    order_status,
    COUNT(order_id) AS total_orders,
    SUM(total_sales_value) AS revenue_loss
FROM shiprocket_dataset
WHERE order_status IN ('Returned', 'Pending', 'Cancelled')
GROUP BY month, order_status
ORDER BY month, revenue_loss DESC LIMIT 10;

''', connection)
```

```python
# Incomplete Orders by Region

pd.read_sql_query('''
SELECT 
    region,
    order_status,
    COUNT(order_id) AS total_orders,
    SUM(total_sales_value) AS revenue_loss
FROM shiprocket_dataset
WHERE order_status IN ('Returned', 'Pending', 'Cancelled')
GROUP BY region, order_status
ORDER BY revenue_loss DESC;

''', connection)
```

```python
# Complete Orders by Region

pd.read_sql_query('''
SELECT 
    region,
    order_status,
    COUNT(order_id) AS total_orders,
    SUM(total_sales_value) AS revenue_loss
FROM shiprocket_dataset
WHERE order_status IN ('Delivered')
GROUP BY region, order_status
ORDER BY revenue_loss DESC;

''', connection)
```

```python
# Complete Orders by Product Category with Revenue

pd.read_sql_query('''
SELECT 
    product_category,
    COUNT(order_id) AS total_completed_orders,
    SUM(total_sales_value) AS total_revenue
FROM shiprocket_dataset
WHERE order_status = 'Delivered'
GROUP BY product_category
ORDER BY total_revenue DESC;

''', connection)
```
